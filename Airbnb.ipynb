** UNivariante Analysis **

```python
# price Distribution 
data['price']
```

```python
# Identifying the outliers in the 
sns.boxplot(data['price'])
```

```python
# step 1 : Calculate Q1 and Q3
Q1=data['price'].quantile(0.25)
Q3=data['price'].quantile(0.75)
#step 2 : Calculated IQR
IQR=Q3-Q1
# step 3 : Datermine bounds
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(lower_bound)
print(upper_bound)
```

```python
# to varify whether IQR works well or not 
df=data[(data['price']>=lower_bound)& (data['price']<=upper_bound)]
df
```

**DATA DRIVEN RECCOMENATIONS**

```python
# Data Driven Recommendations

# 1. Focus on neighbourhoods with higher average prices for premium listings.
high_price_areas = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
print("Neighbourhoods with highest average prices:\n", high_price_areas)
```

```python
# 2. Optimize room types based on demand and price.
room_type_stats = df.groupby('room_type')['price'].mean().sort_values(ascending=False)
print("\nAverage price by room type:\n", room_type_stats)
```

```python
# 3. Target listings with high availability and review counts for marketing.
high_availability_reviews = df[(df['availability_365'] > df['availability_365'].mean()) & 
                              (df['number_of_reviews'] > df['number_of_reviews'].mean())]
print("\nListings with high availability and reviews:\n", high_availability_reviews[['neighbourhood_group', 'room_type', 'price', 'availability_365', 'number_of_reviews']].head())
```

```python
# 4. Consider adjusting minimum nights for better occupancy.
min_nights_corr = corr['minimum_nights'].sort_values(ascending=False)
print("\nCorrelation of minimum nights with other features:\n", min_nights_corr)
```

```python
# 5. Most common room type with highest reviews
most_reviewed = df.groupby('room_type')['number_of_reviews'].mean().sort_values(ascending=False)
print("\nRoom type with highest average reviews:\n", most_reviewed)
```
